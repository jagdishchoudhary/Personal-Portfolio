```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imshow, imread
```

181112288 - COMPUTER VISION ASSIGNMENT 1

```python
photo = imread('Jagdish123321.jpg')
```

SAMPLING

```python
from skimage.transform import downscale_local_mean
factors = 3**np.arange(1, 6)
figure, axis = plt.subplots(1, len(factors), figsize=(20, 6))
for factor, ax in zip(factors, axis):
    image = downscale_local_mean(photo, 
                 factors=(factor, factor, 1)).astype(int)
    ax.imshow(image)
    ax.set_title('$N={}$'.format(image.shape[0]))
```

RESULT : As we increase N (image spatial resolution), the image becomes more unclear

QUANTIZATION

```python
photo = imread('Jagdish123321.jpg')
```

```python
factors = 2**np.arange(1, 6)
figure, axis = plt.subplots(1, len(factors), figsize=(20, 6))
for k, ax in zip(factors, axis):
    bins = np.linspace(0, photo.max(), k)
    image = np.digitize(photo, bins)
    image = (np.vectorize(bins.tolist().__getitem__)
                         (image-1).astype(int))
    ax.imshow(image)
    ax.set_title('$k = {}$'.format(k))
```

RESULT : As k value increases, the image becomes more clear.
